```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set better style
sns.set(style="whitegrid")

plt.figure(figsize=(18, 5))

# Plot 1: All Students vs Low-Income (colored by gradelevel)
plt.subplot(1, 3, 1)
sns.scatterplot(data=df, x='all_students', y='low-income', hue='gradelevel', palette='tab10')
plt.title("All Students vs Low-Income (by Grade Level)")

# Plot 2: All Students vs Students With Disabilities (colored by gradelevel)
plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='all_students', y='students_with_disabilities', hue='gradelevel', palette='Set2')
plt.title("All Students vs Disabilities (by Grade Level)")

# Plot 3: Low-Income vs Students With Disabilities (colored by gradelevel)
plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x='low-income', y='students_with_disabilities', hue='gradelevel', palette='Dark2')
plt.title("Low-Income vs Disabilities (by Grade Level)")

plt.tight_layout()
plt.show()
```
